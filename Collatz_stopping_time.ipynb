```python
import matplotlib.pyplot as plt

def collatz_steps(Cn):
    steps = 0
    while Cn != 1:
        if Cn % 2 == 0:
            Cn //= 2
        else:
            Cn = 3 * Cn + 1
        steps += 1
    return steps


N = []
T = []
for Cn in range(1, 10001):
    N.append(Cn)
    T.append(collatz_steps(Cn))


plt.figure(figsize=(10, 6))
plt.scatter(N, T, s=10)
plt.xlabel("Starting number")
plt.ylabel("Number of steps to reach 1")
plt.title("Collatz Steps for T(Cn) to 1")
plt.show()

```
